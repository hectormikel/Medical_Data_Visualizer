```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

 ## Import Data

```python
# Import data
df = pd.read_csv("medical_examination.csv")
df.head()
```

## Data Cleaning

#### Filtering out patient segments that represent plausibly inaccurate medical data, as recommended:

1. Diastolic pressure is higher then systolic (Keep the correct data with df['ap_lo'] <= df['ap_hi']))
2. Height is less than the 2.5th percentile (Keep the correct data with (df['height'] >= df['height'].quantile(0.025)))
3. Height is more than the 97.5th percentile
4. Weight is less then the 2.5th percentile
5. Weight is more than the 97.5th percentile

```python
df_heat = df[(df['ap_lo'] <= df['ap_hi']) &
            (df["height"] >= df["height"].quantile(0.025)) &
            (df["height"] <= df["height"].quantile(0.975)) &
            (df["weight"] >= df["weight"].quantile(0.025)) &
            (df["height"] <= df["height"].quantile(0.975))]
```

## Data Transformation

### I. Add "overweight" column

```python
df.head()
```

```python
df["overweight"] = (df["weight"]/(df["height"]/100)**2) # BMI formula

df.loc[df["overweight"] <= 25, "overweight"] = 0
df.loc[df["overweight"] > 25, "overweight"] = 1
    
df["overweight"]
```

### II. Normalize data by making 0 always good and 1 always bad. If the value of 'cholestorol' or 'gluc' is 1, make the value 0. If the value is more than 1, make the value 1.

```python
df.head()
```

#### II. I. Normalise "cholesterol" data

```python
df["cholesterol"]
```

```python
df.loc[df["cholesterol"] == 1, "cholesterol"] = 0
df.loc[df["cholesterol"] > 1, "cholesterol"] = 1
df["cholesterol"]
```

#### II. II. Normalise "gluc" data

```python
df["gluc"]
```

```python
df.loc[df["gluc"] ==1, "gluc"] = 0
df.loc[df["gluc"] > 1, "gluc"] = 1
df["gluc"]


# df.loc[df["cholesterol"] == 1, "cholesterol"] = 0
# df.loc[df["cholesterol"] > 1, "cholesterol"] = 1
# df["cholesterol"]
```

## Data Analysis

### III. Draw Categorical Plot

#### Create DataFrame for cat plot using `pd.melt` using just the values from 'cholesterol', 'gluc', 'smoke', 'alco', 'active', and 'overweight'.

```python
df.head()
```

```python
df_cat = pd.melt(df, id_vars = "cardio", value_vars =["active", "alco", "cholesterol", "gluc", "overweight", "smoke"]) 
df_cat
```

#### Group and reformat the data to split it by 'cardio'. Show the counts of each feature. You will have to rename one of the columns for the catplot to work correctly.

```python
# Draw the catplot with 'sns.catplot()'
fig = sns.catplot(data=df_cat, kind="count",  x="variable", hue="value", col="cardio")
fig.set_axis_labels("variable", "total")
```

Onto Heat Map Section (cleaned DataFrame df_heat in Data Cleaning Section)

```python
# Calculate the correlation matrix
corr = df_heat.corr()
corr
```

```python
# Generate a mask for the upper triangle
# mask = sns.heatmap()
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
```

```python
# Set up the matplotlib figure
fig, ax = plt.subplots(figsize=(9,9))
```

```python
# Draw the heatmap with 'sns.heatmap()'
plt.figure(figsize=(14,8))
sns.set_context("paper", font_scale=1)
sns.heatmap(corr, mask=mask, fmt=".1f", vmax=.3, center=0, annot=True,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```
